# RL CartPole: Double DQN

```python
import numpy as np
import gym
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
import random

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
# Just making otf.function to model.predict
from collections import deque
import time

#---------- Environment and seeds -----------------#

env = gym.make('CartPole-v1')
input_shape = env.observation_space.shape
n_actions = env.action_space.n
print("Observation space:", input_shape, "n_actions:", n_actions)

tf.keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)
env.action_space.seed(42)  # Ensures deterministic action sampling

#---------- Parameters -----------------#

batch_size = 256
discount_factor = 0.99
replay_buffer = deque(maxlen=200000)
n_steps = 500
n_episodes = 3000
tau = 0.01


#---------- Neural nets-----------------#

online_model = Sequential([
    Dense(32, activation='elu',input_shape=input_shape),                #Elu activation function better than Relu for negative values (Relu can have dying neurons)
    Dense(32, activation='elu'),
    Dense(n_actions, activation='linear')
])

inital_weights = online_model.get_weights()    

target_model = tf.keras.models.clone_model(online_model)
target_model.set_weights(inital_weights)

# Optimizer and loss
optimizer = Adam(learning_rate=0.01)
loss_fn = tf.keras.losses.MeanSquaredError()


#---------- Neural fast functions-----------------#

@tf.function()
def predict_onlineModel_fast(state):
    return online_model(state, training=False)

@tf.function()
def predict_targetModel_fast(state):
    return target_model(state, training=False)


#---------- Other Functions-----------------#

def epsilon_greedy_policy(state, epsilon=0):
    if np.random.rand() < epsilon:
        return np.random.randint(n_actions)
    else:
        #Q_values = model.predict(state[np.newaxis],verbose= 0)                 #np.newAxis to transfomr teh state from (4,) to (1,4), which is needed as input shape for teh Neural Model to predict
        #print(state)
        #start_time_predict = time.time()
        Q_values = predict_onlineModel_fast(state[np.newaxis])
        #end_time_predict = time.time()
        #print("Time taken to predict: ", end_time_predict - start_time_predict)
        return np.argmax(Q_values[0])                               #Q_values[0] to get the Q_values from the first dimension of the array (for teh first row or only row. Q_values [0] has 2 values, one for each possible action.)


def sample_experiences(batch_size):
    indices = np.random.randint(len(replay_buffer), size=batch_size)
    batch = [replay_buffer[index] for index in indices]
    states, actions, rewards, next_states, dones = [
        np.array([experience[field_index] for experience in batch])
        for field_index in range(5)]
    return states, actions, rewards, next_states, dones

# Play one Step, using Epsilon Greedy Policy and store the state, action, reward, next_state and done in the replay buffer
def play_one_step(env, state, epsilon):
    action = epsilon_greedy_policy(state, epsilon)
    next_state, reward, done, info = env.step(action)
    replay_buffer.append((state, action, reward, next_state, done))
    return next_state, reward, done, info

def play_n_steps(env, state, epsilon, n_steps):
    total_reward = 0 
    for _ in range(n_steps):
        next_state, reward, done, info = play_one_step(env, state, epsilon)
        total_reward += reward  # Accumulate reward
        state = next_state
        if done:
            break
    return state, total_reward, done  

@tf.function()  # This will optimize the function into a computational graph for better performance
def optimize_model(states, target_Q_values, actions, n_actions, model, loss_fn, optimizer):
    # One-hot encode the actions taken
    mask = tf.one_hot(actions, n_actions)
    
    # Perform the forward pass and loss calculation
    with tf.GradientTape() as tape:
        # Get Q-values for the current states
        all_Q_values = model(states)
        
        # Select the Q-values corresponding to the taken actions using the mask
        Q_values = tf.reduce_sum(all_Q_values * mask, axis=1, keepdims=True)
        
        # Compute the loss between target Q-values and predicted Q-values
        loss = tf.reduce_mean(loss_fn(target_Q_values, Q_values))
    
    # Calculate gradients and apply them
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    
    return loss


def training_step(batch_size):
    # Sample a batch of experiences
    experiences = sample_experiences(batch_size)
    states, actions, rewards, next_states, dones = experiences
    #print("rewards shape: ", rewards.shape)
    #print("dones shape: ", dones.shape)

    
    # Step 1: Get Q-values for next states using the online model (model)
    next_Q_values = predict_onlineModel_fast(next_states)

    #print("next_Q_values: ", next_Q_values.shape)

    # Step 2: Select the best actions based on the Q-values from the online model
    best_next_actions = np.argmax(next_Q_values, axis=1)
    #print("best_next_actions: ", best_next_actions.shape)

    # Step 3: One-hot encode the best actions
    next_mask = tf.one_hot(best_next_actions, n_actions).numpy()
    #print("next_mask: ", next_mask.shape)

    # Step 4: Get Q-values for the next states using the target model (target_model)
    next_best_Q_values = predict_targetModel_fast(next_states)
    #print("next_best_Q_values: ", next_best_Q_values.shape)
    
    # Step 5: Calculate the next best Q-values (for the selected actions) from the target model
    next_best_Q_values = tf.reduce_sum(next_best_Q_values * next_mask, axis=1, keepdims=False).numpy()   #Works
    #print("next_best_Q_values: ", next_best_Q_values.shape)
    #next_best_Q_values = (next_best_Q_values * next_mask).sum(axis=1)          # does not work, as tf.function does nto return a tf.Tensor where you can applñy.sum as in numpoy

    # Step 6: Calculate the target Q-values (Bellman equation)
    target_Q_values = rewards + (1 - dones) * discount_factor * next_best_Q_values
    #print(f" Target Q_values: {target_Q_values.shape} = rewards: {rewards.shape} + (1 - dones): {(1 - dones).shape} * discount_factor: {discount_factor} * next_best_Q_values: {next_best_Q_values.shape}")
    
    #print("target_Q_values: ", target_Q_values.shape)
    # Reshape target_Q_values to be (batch_size, 1)
    #target_Q_values = tf.reshape(target_Q_values, (batch_size, 1))
    #print("target_Q_values after reshaping: ", target_Q_values.shape)
    
    # Step 8: Perform backpropagation using the current model to minimize the loss
    loss = optimize_model(states, target_Q_values, actions, n_actions, online_model, loss_fn, optimizer)
    
    




# Reset Gym environment
env = gym.make('CartPole-v1')
tf.keras.backend.clear_session()
env.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)
env.action_space.seed(42)  # Ensures action selection is deterministic
replay_buffer.clear()

online_model.set_weights(inital_weights)
target_model.set_weights(inital_weights)


episode_rewards = []
mean_rewards = []
best_score = 0

start_time = time.time()
for episode in range(n_episodes):
    obs = env.reset(seed=42)
    total_reward = 0
    start_time_episode = time.time()

    if episode > 1000:
        epsilon = max(1 - (episode-1000) / (2500-1000), 0.01)
    else:
        epsilon = 1
    
    '''
    for step in range(500):
        obs, reward, done, info = play_one_step(env, obs, epsilon)
        total_reward += reward  # Accumulate reward
        
        if done:
            break
    '''
    start_play_time = time.time()
    obs, total_reward, done = play_n_steps(env=env, state=obs, epsilon=epsilon, n_steps=n_steps)
    end_play_time = time.time()
    #print("Time taken to play: ", end_play_time - start_play_time)
    episode_rewards.append(total_reward)
    mean_r = np.mean(episode_rewards[-40:]if len(episode_rewards)>40 else episode_rewards)
    mean_rewards.append(mean_r)

    if mean_r > best_score:
        best_weights_online = online_model.get_weights()
        best_weights_target = target_model.get_weights()

        best_score = mean_r

    if episode > batch_size:
        start_train_time = time.time()
        training_step(batch_size)
        end_train_time = time.time()
        #print("Time taken to train: ", end_train_time - start_train_time)

    
    target_model.set_weights(online_model.get_weights())
    target_weights = target_model.get_weights()
    online_weights = online_model.get_weights()
    for index in range(len(target_weights)):
        target_weights[index] = (1-tau) * target_weights[index] + tau * online_weights[index]
    target_model.set_weights(target_weights)


    end_time_episode = time.time()

    print(f"Episode: {episode},Total Reward: {total_reward}, Mean Reward: {mean_r}, epsilon: {epsilon}")
    #print("Time taken for episode: ", end_time_episode - start_time_episode)
end_time = time.time()
print("Time taken to train: ", end_time - start_time)

online_model.set_weights(best_weights_online)
target_model.set_weights(best_weights_target)

# Use Seaborn to plot rewards per episode
sns.set_style("darkgrid")  # Set a nice style
plt.figure(figsize=(10, 5))  # Set figure size

sns.lineplot(x=range(len(episode_rewards)), y=episode_rewards, label="Total Reward per Episode")
sns.lineplot(x=range(len(mean_rewards)), y=mean_rewards,  label="Mean Total Reward per Episode")

plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("Training Progress: Rewards per Episode")
plt.legend()
plt.show()
```
